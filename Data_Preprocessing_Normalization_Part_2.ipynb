# Single Feature Scaling

* **reshape(-1)** - converts the array into a line(row) vector
* **reshape(-1,1)** - converts the array into a column vector
* **reshape(1,-1)** - converts the array back to its original shape 

```python
X = np.array(data['tamponi']).reshape(-1,1)
scaler = MaxAbsScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)
data['single feature scaling'] = X_scaled.reshape(1,-1)[0]
data.head(10)
```

```python
scaler.inverse_transform(X_scaled)
```

# Min Max

```python
X = np.array(data['totale_casi']).reshape(-1,1)
scaler = MinMaxScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)
data['min max'] = X_scaled.reshape(1,-1)[0]
data.head(10)
```

#Z-score

```python
X = np.array(data['deceduti']).reshape(-1,1)
scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)
data['z score'] = X_scaled.reshape(1,-1)[0]
data.head(10)
```
