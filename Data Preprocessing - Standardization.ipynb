```python
#Importing the necessary libraries
import pandas as pd
from scipy.stats import zscore,zmap
import numpy as np
```

```python
#Loading the dataset
data = pd.read_csv('D:\\SLIIT\\3rd year 2nd sem\\Fundamentals of Data Mining\\Coding\\dpc-covid19-ita-regioni.csv')
data.head(10)
```

```python
#Dropping columns with missing values
data.dropna(axis=1,inplace=True)
data.tail(10)
```

# Z Score

```python
data['zscore_deceduti'] = zscore(data['deceduti'])
data['zscore_deceduti']
```

# Z Map

```python
zmap(data['deceduti'],data['terapia_intensiva'])
```

# Detect Outliers

```python
threshold = 2
data['outliers'] = np.where((data['zscore_deceduti'] - threshold > 0), True, np.where(data['zscore_deceduti'] + threshold < 0, True,False))
```

```python
data.drop(data[data['outliers'] == True].index,inplace=True)
```
